# DLAI S08A04 - Working with null values

```
"""
In DataScience,
One of the common issue we face is --> the dataset will have the missing values. 

So, lets understand as how we deal with missing values in the pandas dataframe.

For our learning purpose, I have modified my existing dataset,
So this time, I will load the dataset as :
"""
import pandas as pd
df = pd.read_csv("new_olympic.csv", skiprows = 1, index_col= "Country")
df.head()
```

```
"""
So, this has loaded the dataset.

Now, lets call the info() method on our dataframe.
So, it is
"""
df.info()
```

```
"""
We know that, there are 152 entries in our dataframe,
and we can observe that --> we have an entry missing in column 'No. of Summer Games' as this has 151 entries, 
and in the same way, we have 3 entries missing in column ' Summer Total".

So that means, we have entries missing in some of the columns.

Whenever we come across a situation like this, that is --> our dataset having missing values in it

We can deal with them in two ways:
- First option is, delete a specific rows or columns which has 'null' value

- and, the second option is, Replacing the 'null' with a real value using a technique
known as imputation

Before proceeding to directly deal with 'null' values, First, lets see how many 'null' values, that is 'missing values'
in each column of our dataset. 
To check the 'null' values for each cell, we use the method called 'isnull()' 
So, we write as :
"""
df.isnull()
# I will then execute the code
```

```
"""
Okay,
As we can observe in the output, the 'isnull' method has returned the dataframe 
with values as 'True' or 'False' depending on that cell’s value.

But, this is not much helpful right, we dont know how many 'null' values are present in each column,
hence we use this method in combination with 'sum' method to count the number of nulls in each column.
So, I will write as:
"""
df.isnull().sum() 
```

```

"""
Okay, 
We can now see that, how many entries are missing in each column
So, in this way we return the total number of null values in each column.


Now, lets look at the first method of dealing with 'null' values,That is
deleting the row which has at least a single 'null' value.
To do that, we make use of method, 'dropna()'
So, I will call as :
"""
df.dropna()
```

```
"""
Okay, 
if we scroll down and observe here at the end of dataframe,
We can see that, the shape of our dataframe after applying the method --> 'dropna()' is (140,15)

that means, we have 140 rows and 15 columns
So, we have dropped around 12 rows of data which had atleast a single 'null' values.

Now if you observe, in all the methods which we had applied till now,
Just by applying the methods on the dataframe, doesnot change the original
dataframe.
TO SHOW THE SAME, I WILL DISPLAY THE SHAPE OF OUR DATAFRAME,
"""
df.shape
```

```
"""
We still have the shape of dataframe as (152,15) , and dropna() method which we just applied
did not had any affect on the original dataframe.

Now, If we want the changes to apply on the original dataframe, then we have to specify by making use
of the parameter, 'inplace'
So, this time, I will write as :
"""
df.dropna(inplace= True)
```

```
"""
We can see that, the shape of our dataframe is --> 140,15
That means, the changes we applied using the dropna() method have been saved into the dataframe.

As you have observed, when we call the 'dropna()' method, 
the rows which has atleast a single value has been deleted by default.

Instead of rows, we can also delete the respective columns which has atleast a single
'null' value by specifying the axis.

To illustrate the same, I will again create our original dataframe with null values.
So,
"""
df = pd.read_csv("new_olympic.csv", skiprows = 1, index_col= "Country")
df.head()
```

# DLAI S08A06 - Imputation

```
"""
In this video,
Lets understand how we can perform imputation on our dataframe.

This is a technique which we use when we are dealing with null values in our dataframe.

We employ this technique when we have large chunk of data with null values.
In such cases, If we directly delete a particular column or a row due to null values,
then, that would result in removing large chunk of data itself.

Hence in such cases, we will fill the null values with mean of the values in that column, so that, 
We will not have any affect after performing the imputation

To shouw you how imputation works, lets first import our data.
So,
"""
import pandas as pd
df = pd.read_csv("new_olympic.csv",skiprows= 1, index_col="Country")
df.head()
```

```
"""
Okay, we have now created the dataframe. Now, lets call the info() method.
"""
df.info()
```

```
"""
Okay,
For the ease of learning, lets consider the column which has the data type of numbers

So,
lets choose the column --> No. Combined
As we can see, this has 150 rows of data, That means, we have 2 rows missing in our dataframe,
and the data type of that column is --> float64

Now, as a process of imputation, lets fill the null values with mean values of that columns.

As a first step, I will compute the mean for that column values
"""
total = df['No. Combined']
import numpy as np
mean_total = np.mean(total)
```

```
mean_total
```

```
"""
As we can see in the output,
The mean of the column --> 'No. Combined' is 33.4172 RERECORD AS 20.78
Now, lets fill the null values with the mean value - 33.4172 RERECORD AS 20.78

For that, we make use of 'fillna()' method
I will write as:
"""
df['No. Combined'].fillna(mean_total , inplace = True )
df.info()
```

```
"""
As we can see in the output, 
We have now filled the null values of 'No. Combined' with mean values.

The way we have made use of the fillna() method is,

We apply this method on a particular column, and inside the parentheses of the fillna() method,
as a first argument we specify the value to be filled on the column, and as a second optional parameter, I have 
specified as 'inplace' = True

That means, the dataframe will be changed inplace.
As a result, all the null values of that particular column has been replaced with mean value of that column.

So, This completes our learning on Pandas Library,
We will make use of all the concepts which we had learnt till now in the upcoming
section on --> 'Project on Machine learning', Completing that project will help you get hands-on experience as how each concept works

In the next section,
We are going to learn as how to work with Matplotlib library.
See you then.
"""
```
