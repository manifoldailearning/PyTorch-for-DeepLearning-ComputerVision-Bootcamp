# DLAI S09A02 - Format the Plot

```
"""
In this video,
Lets continue our learning on formatting our matplotlib plot

We are now in documentation page,
Here, Under the heading 'Markers',
We have the different types of markers supported by matplotlib.
We have --> 	point marker , pixel marker ,	circle marker and so on.
These are the character, that we write in our code to choose the required marker style.

To illsutrate as how markers looks like, and how to include in our code,
I will write the code like this.

So , First, I will import the library,
"""
import matplotlib.pyplot as plt
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,marker = '.', label = "Sine Function")
plt.legend()
plt.show()
```

```
"""
For our learning purpose,
I have included the character of 'pointer' for the parameter --> 'marker'
inside the plt.plot() function

hence, we have the pointer type markers in our plot.

Now, lets make use of one more marker style,
This time, lets create a plot with marker style as - Diamond
So, the symbol is --> 'D' 

So this time, I will write as
"""
plt.plot(x,y,marker = 'D', label = "Sine Function")
plt.legend()
plt.show()
```

```
"""
As we can observe,
our plot is now having the markers in the shape of diamonds.

Now, lets use the marker 'o' , that is --> circle marker
We also refer this marker as --> scatter plot marker 
So, i will write as :
"""
plt.plot(x,y, marker = 'o')
plt.show()
```

```
"""
Okay, we have the markers in the form of circles.

I suggest you to run the commands like this to all the other marker style supported by matplotlib to
get an idea as how each marker style looks like

Now, lets go back to our documenation page,

After the markers,
We have the heading --> 'line styles'
Under the heading, we have the table which provides the details of all the different line styles supported by matplotlib
We have -->
 - 	solid line style
 -	dashed line style
 -	dash-dot line style , and
 -	dotted line style 

We also have the respective characters which we specify to get that particular line style.

To illustrate, how we make use of line style,
I will make use of dashed line style to my plot.
So, I will write as :
"""
plt.plot(x,y, linestyle = '--' , label = "Sine Function")
plt.legend()
plt.show()
```

```
"""
As we can see in the plot,
Instead of solid line which we had before, our plots are now having the dotted lines.
To do that,
We just specified the respective character to the parameter --> linestyle

moving on, lets change the plot with linestyle as 'dash dot'
So, I will write as:
"""
plt.plot(x,y, linestyle = '-.' , label = "Sine Function")
plt.legend()
plt.show()
```

```
"""
okay.
This has created the plot with the linestyle of --> dashed dot

In this way, You can play with different line style in your plot .
(pause 2 seconds)
Now, lets go back to the documentation page,

The next heading we have is, Colors.
Yes. We can also specify the color of our line by making use of these characters.
The table shows character code for each color, We have
'b' for	blue
'g' for	green
'r' for	red
'c' for	cyan
'm' for	magenta , and so on

Lets include the same to our plot to build intuition as how to use these,
So, I will write as :
"""
plt.plot(x,y, linestyle = '-.' , color = 'g', label = "Sine Function")
plt.plot(x,z, linestyle = '--' , color = 'r', label = "Cosine Function")
plt.legend()
plt.show()
```

```
"""
So this time,
I have created two plots in a single figure, 
1st is Sine function with line style of 'dash dot' and color as 'green' , and the 2nd is cosine
function with line style as 'dash' and color as 'red'

in this way,
we make use of formatting to customize our plot.

Now,
There is one trick i want to show.

Lets go back to our documentation page.

Here,
Observe the box under : 'Example format strings'

it has suggested various ways of mentioning the format of our plot.

That means,
Instead of specifying marker, linestyle and color as separate argument,
We can call them in a single argument.

if we simply use the character 'b', our plot will have blue markers with default shape
and the usage of character 'or'  will result in plot of red circles, which is nothing but a scatter plot
the usage of character '-g'   will result in plot of green solid line
Similary, the usage of character '--'   will result in dashed line with default color
and the usage of this symbol ('^k:')  will result in black triangle_up markers connected by a dotted line

To build the intuition as how each plot would look like, I will use the same in my plots 

Okay,
First I will make use of character 'b'. So i will write as
"""
plt.plot(x,y,'b')
plt.show()
```

```
"""
So, we have the got the line in blue color
Now, lets use character 'or'
(dont read the code)
"""
plt.plot(x,y,'or')
plt.show()
```

```
"""
Okay, this has created the plot with red circles
This is nothing but a scatter plot.

Now, I will use the character --> '-g'
(dont read the code)
"""
plt.plot(x,y,'-g')
plt.show()
```

```
"""
This has created the plot with solid green line
Now, i will use the character '--' (double dash)
(dont read the code)
"""
plt.plot(x,y,'--')
plt.show()
```

```
"""
Now,
Apart from adding the labels, and legends for our plot,
We can also add the title to our plot

For that, we make use of plt.title() method.
So, I will include the title like this :
(dont read the code)
"""
plt.plot(x,y,'-b', label = "Sine Function")
plt.title("Sine Function heading")
plt.legend()
plt.show()
```

```
"""
Okay.
As you can see in the plot,
We have the heading printed by making use of --> plt.title() method

Now, apart from this,
We can also make use of --> plt.xlim() and plt.ylim() methods to specify the x & y axis range

As a default,
We can see that,
X axis values ranges between 0 and 10, and
Y axis values ranges between -1 and 1

now, I will change the x axis and y axis range like this
(dont read the code)
"""
plt.plot(x,y,'-b', label = "Sine Function")
plt.title("Sine Function heading")
plt.xlim(-10,10)
plt.ylim(-5,5)
plt.legend()
plt.show()
```

# DLAI S09A04 - Hist plot

```
"""
In this video,
Lets learn how to create a histogram plot.

Guys. This is one of the very important plot in Datascience.
Because, in most cases, this is the first plot we are going to use on a realworld dataset
to get a visual understanding as how data is distributed

To show you how  a simple histogram looks like.

I will define the variable 'x' like this
"""
x = np.random.randn(1000)
# So, this will generate the 1000 random numbers . 
# Now using 'x' , I will create the histogram plot as

plt.hist(x)
#and then,
plt.show()
```

```
"""
This is how an histogram plot looks like.
To create a histogram plot, we simply pass the data inside the : plt.hist() function

The statistical reason to use the histogram plot is, it will help us to understand the -
frequency distribution of a continuous data.
(pause 2 seconds)

To help you build more inutition on this plot,
This time, I will create a new data as.

"""
data = np.array([1,1,2,2,2,3,3,3,5,5,6,7,8,8,8,8,9,9,9,9,9])
```

```
"""
So, in this data, we have the number '1' --> twice,
that means, the frequency of number 1 is 2

Similarly, frequency of number 2 is three,
the frequency of number 3 is also 3, 
and so on.

(pause a second)
We know how the data looks like.

Now, Lets create the histogram plot.
I will write as -->
"""
plt.hist(data)
plt.show()
```

```
"""
Okay,

Since the frequence of number 1 is equal to 2, we have a histogram plot represeting the value:2 here,
and for the number 2 & 3, the frequency is 3,
Hence we have the plot represeting the frequency of 3

Since, We do not have any value for  '4' , it is blank against the number : 4

I hope you now got an idea as what we are trying to represent in the histogram plot.

(pause 5 seconds)

This plt.hist() function accepts other parameters, using which, we can customize the plot

one of the most used parameter inside the hist() function is, 'bins'

Bins are the blocks that represent the data, and We use it, to combine the values before getting the frequency. 

To illustrate the use of bins, I will write the code like this.
"""
plt.hist(data, bins = 3)
plt.show()
```

```
"""
This time, I have set the value of parameter --> 'bins' to be equal to 3,

Hence we have got the histogram plot after combining the three values.

To explain the plot,
the data, the frequence of 1 is 2, and the frequence of 2 & 3 is 3 
If we combine it will be : 2 + 3 + 3 , which is equal to 8

Hence,
We have the historgram plot that shows frequency of 8 in the first bin.

Similary, 
If we combine the values of 4,5 & 6, it is equal to 3. Hence we have the histogram plot
that shows the value 3 for the second bin.

To Summarize,
By specifying the bin size, we can control as how values to be combined before getting the frequency.

(pause 2 seconds)
In Data Science,
the most common way of using the histogram plot, is with pandas dataframe.

To illustrate the creation of histogram plot using pandas dataframe,
I will first create a pandas dataframe from a csv file.
"""
import pandas as pd
df = pd.read_csv("haberman.csv")
#okay, lets display the first 5 entries of our dataframe
df.head()
```

```
"""
So, a dataframe has been created.

Now observe this very carefully

Whenever we are creating the histogram plot for a dataframe,
We will call the 'hist()' function like this
"""
df.hist()
plt.show()
```

```
"""
As we can see in the output,
We have the histogram plot for each of the columns present in the dataframe.


Here, observe the plot. The figure size of our plot is very small.
To increase the readability, we can specify the required figure size. 

So, I will write as:
"""
df.hist(figsize=(20,15))
```
